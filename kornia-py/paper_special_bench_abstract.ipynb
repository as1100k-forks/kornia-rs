```python
import matplotlib.pyplot as plt
import numpy as np

# 4 operations => 4 subplots
fig, axes = plt.subplots(2, 2, figsize=(8, 8))

# Sample data
# from benchmark
labels = ["480p", "720p", "1080p", "1440p", "4k"]
kornia_times = [142.35, 286.58, 752.17]
image_times = [2212.5, 3940.6, 9051.7]

x = np.arange(len(labels))  # label locations
width = 0.2  # width of the bars

fig, plt = plt.subplots(figsize=(8, 6))

# Plotting the bars
rects1 = plt.bar(x - width/2, kornia_times, width, label='Kornia')
rects2 = plt.bar(x + width/2, image_times, width, label='Image')

# Add labels, title and custom x-axis tick labels
plt.set_xlabel('Resolution')
plt.set_ylabel('Processing Time (µs)')
plt.set_title('Horizontal Flip (u8, RGB, CPU, Intel i7-13700H)')
plt.set_xticks(x)
plt.set_xticklabels(labels, rotation=30, ha='right')
plt.legend()

# Optional: add value labels on bars
for rect in rects1 + rects2:
    height = rect.get_height()
    plt.annotate(f'{height:.1f}',
                xy=(rect.get_x() + rect.get_width() / 2, height),
                xytext=(0, 3),  # offset text upward
                textcoords="offset points",
                ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
